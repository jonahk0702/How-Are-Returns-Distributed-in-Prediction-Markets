```python
def hex_to_decimal(hex_string):

    # Remove the '0x' prefix if it exists
    if hex_string.startswith("0x"):
        hex_string = hex_string[2:]
    
    # Convert to decimal
    decimal_number = int(hex_string, 16)
    return decimal_number
```

```python
import hashlib

def keccak256(function_signature: str) -> str:
    """Compute the Keccak-256 hash of a function signature and return the first 4 bytes as hex."""
    hash_bytes = hashlib.new('sha3_256', function_signature.encode()).digest()
    return hash_bytes[:4].hex()

functions = [
    "addAdmin(address)",
    "admins(address)",
    "collateral()",
    "ctf()",
    "exchange()",
    "isAdmin(address)",
    "matchOrders((uint256,address,address,address,uint256,uint256,uint256,uint256,uint256,uint256,uint8,uint8,bytes),((uint256,address,address,address,uint256,uint256,uint256,uint256,uint256,uint256,uint8,uint8,bytes)[]),uint256,uint256[],uint256)",
    "onERC1155BatchReceived(address,address,uint256[],uint256[],bytes)",
    "onERC1155Received(address,address,uint256,uint256,bytes)",
    "removeAdmin(address)",
    "renounceAdmin()",
    "withdrawFees(address,uint256,uint256)"
]

input_selector = "d2539b37"  # Extracted from the input

# Find the matching function
for function in functions:
    selector = keccak256(function)
    if selector == input_selector:
        print(f"Match found: {function}")
        break
else:
    print("No match found.")
```

```python
import hashlib

# Function signature to hash

def dodo(function_signature):
    

    # Keccak-256 hashing (Ethereum's hashing algorithm)
    keccak_hash = hashlib.new('sha3_256', function_signature.encode()).hexdigest()

    # Extract the first 4 bytes (8 characters)
    selector = keccak_hash[:8]

    print(f"Function Signature: {function_signature}")
    print(f"Selector: {selector}")
```
